```python
import pandas as pd
```

```python
df = pd.read_excel("solutions.xlsx")
df
```

```python
columns = ["accuracy", "NLL", "ECEL", "SINV", "T1agree", "KLDiv"]
for col in columns:
    if col in ["T1agree", "KLDiv"]:
        df[col] = df[col].transform(lambda x: x if x != "na" else None)
    mean = df.groupby(["Model","temp"])[col].transform("mean")
    sensitivity = df.groupby(["Model", "temp"])[col].transform(lambda x: x.max()- x.min())
    df[col] = mean
    df[col + " sensitivity"] = sensitivity
```

```python
df = df.drop_duplicates(subset=['Model', 'temp'])
df = df.reset_index(drop=True)
```

```python
df
```

```python
cols = ["T1agree", "KLDiv", "T1agree sensitivity", "KLDiv sensitivity"]
for col in cols:
    df[col] = df[col].fillna("na")
df
```

```python
df.to_excel("processed_results.xlsx")
```
